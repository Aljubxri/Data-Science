2a.	A bar chart showing the profitability of the film. The X axis is the film. The Y axis is the profitability. 
The bars should be sorted from the most to the least profitable 


```python
df = pd.read_csv(r'C:\Users\aljub/Desktop/School/CSC 4730/Project 2/HollywoodsMostProfitableStories.csv')

df = df.sort_values(by='Profitability', ascending=True)

trace1 = go.Bar(x=df['Film'], y=df['Profitability'])

layout1 = go.Layout(title='Film vs Profitability', plot_bgcolor='rgb(230, 230,230)',
                    yaxis=go.layout.YAxis(
                        title=go.layout.yaxis.Title(
                        text='Profitability')
                    )
                   )

fig1 = go.Figure(data=[trace1], layout=layout1)

iplot(fig1, filename='HollywoodsMostProfitableStoriestable')
```

3c.A line chart with four lines: price1998, price2007, price2011, and price2014. The X axis is the house num. 
    The Y axis is the value from the four columns listed above. 
    When the mouse cursor hovers over a marker, the street address of the house should be displayed in the tooltip.

4b.	Create a grid with four cells. In the first row, show two charts: a histogram showing the number of menâ€™s champions for different countries and histogram showing the number of womenâ€™s champions for different countries. 
In the second row, show two charts: a histogram showing the number of menâ€™s runners-up for different countries and histogram showing the number of womenâ€™s runners-up for different countries. 



```python
from collections import Counter

data = pd.read_csv(r'C:\Users\aljub/Desktop/School/CSC 4730/Project 2/wimbledons_champions.csv')

m_champ_nationality=data['ChampionNationality'].head(n=130)
f_champ_nationality=data['ChampionNationality'].tail(n=131)
f_runner_nationality=data['Runner-up Nationality (Men\'s)'].tail(n=123)
m_runner_nationality=data['Runner-up Nationality'].head(n=130)


male_champions = Counter(m_champ_nationality)
female_champions = Counter(f_champ_nationality)
female_runner = Counter(f_runner_nationality)
male_runner = Counter(m_runner_nationality)

fig = plt.figure(figsize=(15.0,15.0))
fig.suptitle('Histograms of Champions and Runner-Ups', fontsize=20)

ax1 = fig.add_subplot(221)
x=male_champions.keys()
y=male_champions.values()
ax1.bar(x,y, width = 1)
plt.ylabel('Wins (Male Champions)')
plt.xlabel('Country (Male Champions)')

ax2 = fig.add_subplot(222)
x1=female_champions.keys()
y1=female_champions.values()
ax2.bar(x1,y1, width = 1)
plt.ylabel('Wins (Female Champions)')
plt.xlabel('Country (Female Champions)')

ax3 = fig.add_subplot(223)
x2=male_runner.keys()
y2=male_runner.values()
ax3.bar(x2,y2, width = 1)
plt.ylabel('Runner-Up (Male)')
plt.xlabel('Country (Male Runner-Up)')

ax4 = fig.add_subplot(224)
x4=female_runner.keys()
y4=female_runner.values()
ax4.bar(x4,y4, width = 1)
plt.ylabel('Runner-Up (female)')
plt.xlabel('Country (female Runner-Up)')


plt.tight_layout()
fig = plt.gcf()
```
